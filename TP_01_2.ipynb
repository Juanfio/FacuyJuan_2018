```python
import networkx as nx
import pandas as pd
import matplotlib.pylab as plt
import numpy as np

def ldata(archive):
    f=open(archive)
    data=[]
    for line in f:
        line=line.strip()
        col=line.split()
        data.append(col)
    return data

dolphins = ldata ("dolphins.gml")

g = nx.Graph ()

for i in range (5):
    g.add_nodes_from (dolphins[i])
    g.add_edge (dolphins[i],dolphins[i+1])
    
nx.draw(g,node_size = 40,with_labels=False, font_weight='bold')
plt.show ()
```

```python
import networkx as nx
import pandas as pd
import matplotlib.pylab as plt
import numpy as np

g = nx.read_gml('dolphins.gml')  #lee el gml

plt.subplot(221) #los subplot tienen que ir antes del nx.draw
nx.draw_random(g, node_size = 40,with_labels=False, font_weight='bold')

plt.subplot(222)
nx.draw_circular(g, node_size = 40,with_labels=False, font_weight='bold')

plt.subplot(223)
nx.draw_spectral(g,node_size = 40,with_labels=False, font_weight='bold')

plt.subplot(224)
nx.draw_shell(g, node_size = 40,with_labels=False, font_weight='bold')


plt.show ()
#nx.draw_circular(g, node_size = 40,with_labels=False, font_weight='bold') #dibuja el grafo almacenado en la variable g
#plt.subplot(225)
```
