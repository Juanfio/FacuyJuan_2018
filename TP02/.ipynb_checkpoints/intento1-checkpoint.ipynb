```python
import numpy as np
import matplotlib.pylab as plt
import networkx as nx
import pandas as pd
import random

def ldata(archive):
    f=open(archive)
    data=[]
    for line in f:
        line=line.strip()
        col=line.split()
        data.append(col)
    return data
    
#grafico de nodos
def graficar(archive):

    data=ldata(archive)
    g = nx.Graph ()
    g = g.to_undirected()
    
    for i in range (len(data)):
        g.add_nodes_from (data[i])
        g.add_edge (data[i][0],data[i][1])
```

```python
def grafo (archive):
    data=ldata(archive)
    g = nx.Graph ()
    g = g.to_undirected()

    for i in range (len(data)):
        g.add_nodes_from (data[i])
        g.add_edge (data[i][0],data[i][1])
    return g

#La V es de vulnerabilidad.
gAP = grafo ('tc02Data/yeast_AP-MS.txt')
gLIT = grafo ('tc02Data/yeast_LIT.txt')
gY2H = grafo ('tc02Data/yeast_Y2H.txt')

Lit_Reg = ldata ('tc02Data/yeast_LIT_Reguly.txt')
Litreg = []

for i in range(1,len(Lit_Reg)):
    Litreg.append([Lit_Reg[i][0],Lit_Reg[i][1]])

gLITr = nx.Graph ()
gLITr = gLITr.to_undirected()

for i in range (len(Litreg)):
    gLITr.add_nodes_from (Litreg[i])
    gLITr.add_edge (Litreg[i][0],Litreg[i][1])
```

```python
def impacto_grado (g,largo_red): #largo_red=400
    x = []
    y = []
    #Tomo la componente gigante de mi grafo. Como criterio se toma el largo (key = len) y tomo el subgrafo con el máximo.
    comp_gig = max (nx.connected_component_subgraphs(g) , key = len)
    N = int (comp_gig.number_of_nodes())
    i = 0
    while len (comp_gig) > largo_red: #Modificando el numero podemos determinar hasta cuando va a analizar el largo de la red.
        nodos = [] #Nodos de la componente gigante.
        grados = [] #Grado de ellos.
        for node in  comp_gig.nodes:
            nodos.append (node)
            grados.append (comp_gig.degree(node))
        indice_nodo_grado_mayor = grados.index(max(grados))
        comp_gig.remove_node(nodos[indice_nodo_grado_mayor])
        i = i + 1
        x.append (i/float (N))
        comp_gig = max (nx.connected_component_subgraphs(comp_gig) , key = len)
        y.append(len(comp_gig)/float (N)) #IMPORTANTE:El numerador es el largo NUEVO de la componente gigante y
                                          #el denominador te da el largo antes de la extracción del nodo central.
        
    todo = [x,y]
    return todo
```

```python
def impacto_random (g,largo_red,iteraciones):
    todo=[]
    for _ in range(iteraciones):
        x = []
        y = []
        #Tomo la componente gigante de mi grafo. Como criterio se toma el largo (key = len) y tomo el subgrafo con el máximo.
        comp_gig = max (nx.connected_component_subgraphs(g) , key = len)
        N = int (comp_gig.number_of_nodes())
        i = 0
        for _ in range(largo_red):
#         while len (comp_gig) > largo_red: #Modificando el numero podemos determinar hasta cuando va a analizar el largo de la red.
            nodos = [] #Nodos de la componente gigante.
            grados = [] #Grado de ellos.
            for node in  comp_gig.nodes:
                nodos.append (node)
                grados.append (comp_gig.degree(node))
            nodo = random.choice(nodos) #elijo nodo aleatoriamente
            comp_gig.remove_nodes_from([nodo])
            i = i + 1
            x.append (i/float (N))
            comp_gig = max (nx.connected_component_subgraphs(comp_gig) , key = len)
            y.append(len(comp_gig)/float (N)) #IMPORTANTE:El numerador es el largo NUEVO de la componente gigante y#el denominador te da el largo antes de la extracción del nodo central.
   
        todo.append([x,y])
        
    xx=[]
    yy=[]
    for i in range(len(todo[0][0])):
        xx.append(0)
        yy.append(0)
        
    for i in todo:
        for j in range(len(i[0])):
            xx[j]=i[0][j]/len(todo)+xx[j]
            yy[j]=i[1][j]/len(todo)+yy[j]          
            
    return [xx,yy]
```

```python
def asignar_esencialidad(g):  
    escen = ldata ('tc02Data/Essential_ORFs_paperHe.txt')
    escenciales=[]
    for j in range(len(escen)):
        if len(escen[j])>1:
            escenciales.append(escen[j][1])
#     print(escenciales)    
    e=[]
    for node in g.nodes:
        if node in escenciales:
            g.node[node]['escencialidad']='escencial'
            e.append(node)
        else:
            g.node[node]['escencialidad']='no escencial'
#     print(e)
    return g
```

```python
def impacto_escencial(g):
    g = asignar_esencialidad(g)
    nodos_escenciales=[]
    for node in g.nodes:
        if g.nodes[node]['escencialidad']=='escencial':
            nodos_escenciales.append(node)
    x = []
    y = []
    #Tomo la componente gigante de mi grafo. Como criterio se toma el largo (key = len) y tomo el subgrafo con el máximo.
    comp_gig = max (nx.connected_component_subgraphs(g) , key = len)
    N = int (comp_gig.number_of_nodes())
    i = 0

    while len (nodos_escenciales) > 1: #Modificando el numero podemos determinar hasta cuando va a analizar el largo de la red.
        nodos = [] #Nodos de la componente gigante.
        grados = [] #Grado de ellos.
        nodos_escenciales = []
        for node in  comp_gig.nodes:
            nodos.append (node)
            grados.append (comp_gig.degree(node))
            if g.nodes[node]['escencialidad']=='escencial':
                nodos_escenciales.append(node)
        nodo = random.choice(nodos_escenciales) #elijo nodo escencial aleatoriamente
        comp_gig.remove_nodes_from([nodo])
        i = i + 1
        x.append (i/float (N))
        comp_gig = max (nx.connected_component_subgraphs(comp_gig) , key = len)
        y.append(len(comp_gig)/float (N)) #IMPORTANTE:El numerador es el largo NUEVO de la componente gigante y
                                              #el denominador te da el largo antes de la extracción del nodo central.
        
    todo = [x,y]
    return todo
```

```python
#REMOVER ESCENCIALES
def escenciales(g):
    g = asignar_esencialidad(g)
    nodos=[]
    for node in g.nodes:
        nodos.append(node)
    N=len(g)
    i=0
    
    for node in nodos:
        if g.nodes[node]['escencialidad'] == 'escencial':
            i=i+1
            g.remove_nodes_from([node])
    comp_gig = max(nx.connected_component_subgraphs(g), key=len)
    N1=len(comp_gig)/N
    N2=i/N
    return [N2,N1]
```

```python
def impacto_eigenvector (g):
    x = []
    y = []
    #Tomo la componente gigante de mi grafo. Como criterio se toma el largo (key = len) y tomo el subgrafo con el máximo.
    comp_gig = max (nx.connected_component_subgraphs(g) , key = len)
    N = int (comp_gig.number_of_nodes())
    i = 0
    for _ in range()
    #while len (comp_gig) > 400: #Modificando el numero podemos determinar hasta cuando va a analizar el largo de la red.
        nodos = [] #Nodos de la componente gigante.
        eigenvector = [] #Grado de ellos.
        for node in  comp_gig.nodes:
            nodos.append (node)
            eigenvector.append (nx.eigenvector_centrality_numpy(comp_gig)[node])
        indice_nodo_autovector_mayor = eigenvector.index(max(eigenvector))
        comp_gig.remove_node(nodos[indice_nodo_autovector_mayor])
        i = i + 1
        x.append (i/float (N))
        comp_gig = max (nx.connected_component_subgraphs(comp_gig) , key = len)
        y.append(len(comp_gig)/float (N)) #IMPORTANTE:El numerador es el largo NUEVO de la componente gigante y
                                          #el denominador te da el largo antes de la extracción del nodo central.
        
    todo = [x,y]
    return todo
```
